```python
import pickle
from modal import enable_output
from modalfold import app
from modalfold.esmfold import ESMFold, ESMFoldOutput

# files = ['ASYN.fasta', 'CD28.fasta', 'P53.fasta', 'SUMO.fasta']
files = ['P53.fasta']

sequences = []
for file in files:
    with open(file, "r") as f:
        sequences.append(f.readlines()[1].strip())

with app.run():
    model = ESMFold(output_pdb=True)
    for file, sequence in zip(files, sequences):
        result = model.fold.remote(sequence)
        
        with open(file.replace(".fasta", ".pdb"), "w") as f:
            f.write(result.__dict__['pdb'][0])

        # remove the pdb and save the entire rest of the dict into an object (the values are numpy arrays, so make sure it all works)
        result.__dict__.pop('pdb')
        
        # save the result into a file
        
        with open(file.replace(".fasta", ".pkl"), "wb") as f:
            pickle.dump(result, f)

```
