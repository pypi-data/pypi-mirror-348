```python

def update_plot_kwargs(index, **kwargs):
    """ 

    Returns:
        
    """
    plot_kw = ["label", "linestyle", "linewidth", "color"]
    loc_kwargs = kwargs.copy()
    for key,value in loc_kwargs.items():
        is_list = isinstance(value,list)
        if key in plot_kw and is_list:
            loc_kwargs[key] = value[index]
            print(key, value[index])
    return loc_kwargs
```

```python
print(update_plot_kwargs(2,label=["AA","BB","CC"],linestyle=['-','--','-.'],linewidth=[1,2,3],color=['r','g','b']))
```

```python
import enum
from enum import StrEnum
class aENUM_plotKW(StrEnum):
    LABEL = "label", 
    LINESTYLE = "linestyle", 
    LINEWIDTH = "linewidth", 
    COLOR ="color"

a= [x.value for x in aENUM_plotKW]
print(a) 
```
